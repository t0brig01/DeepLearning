```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np #calculate stuff
import seaborn as sns #visualization
```

```python
df = pd.read_csv('Filtered.csv')
```

```python
from pandas.plotting import scatter_matrix
plt.xkcd()
scatter_matrix(df.loc[:,'TripDuration':'HourNum'], alpha=0.2 , figsize=(15,15) , diagonal='kde')
plt.show()
```

```python
sns.set(palette="RdBu")
df.loc[:,'TripDuration':'HourNum'].hist(bins=15,figsize=(20,15))
plt.show()
```

```python
sns.set(style="white")
df.loc[:,'TripDuration':'HourNum'].boxplot(figsize=(20,15),fontsize=36)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
corrmatrix = df.loc[:,'TripDuration':'HourNum'].corr()
sns.heatmap(corrmatrix, xticklabels=corrmatrix.columns, yticklabels=corrmatrix.columns)
```
